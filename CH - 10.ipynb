## Intoduction to MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,8])
y = np.array([0,250])
plt.plot(x,y)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, '+:r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'k')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 20, linestyle=''or'')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.plot(x,y)

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,15])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("plot1")
plt.xlabel("xlabel1")
plt.ylabel("ylabel1")
plt.grid()

x = np.array([0,1,2,3])
y = np.array([15,25,35,65])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("plot2")
plt.xlabel("xlabel2")
plt.ylabel("ylabel2")
plt.grid()

plt.suptitle("Multiple Plots")
plt.tight_layout()
plt.show()
```
